```python
from transformers import DonutProcessor, VisionEncoderDecoderModel
from PIL import Image
import torch

# Устройство: GPU или CPU
device = "cuda" if torch.cuda.is_available() else "cpu"

# Модель и процессор
model_name = "naver-clova-ix/donut-base"  # или "Nyaaneet/donut-base-ru" для русского
processor = DonutProcessor.from_pretrained(model_name)
model = VisionEncoderDecoderModel.from_pretrained(model_name).to(device)
model.eval()

# Изображение
image = Image.open("../resources/images/4.jpg").convert("RGB")

# Подготовка входов
pixel_values = processor(image, return_tensors="pt").pixel_values.to(device)

# Prompt для HTML-таблицы
prompt = "<s>"  # Или просто "<s>" если модель не дообучена под HTML

decoder_input_ids = processor.tokenizer(prompt, add_special_tokens=False, return_tensors="pt").input_ids.to(device)

# Генерация
with torch.no_grad():
    output = model.generate(
        pixel_values,
        decoder_input_ids=decoder_input_ids,
        max_length=1024,
        pad_token_id=processor.tokenizer.pad_token_id
    )

# Расшифровка
generated_text = processor.batch_decode(output, skip_special_tokens=True)[0]
print("📋 HTML результат:\n")
print(generated_text)
```
